```python
import requests
from bs4 import BeautifulSoup
import re
url = 'https://www.rateinflation.com/consumer-price-index/germany-historical-cpi/'
```

```python
content = requests.get(url=url)
soup = BeautifulSoup(content.content, features="html.parser")
data_body = soup.find_all('tbody', attrs={'class': 'css-1rkiavl eyyd7td2'})
```

```python
print(str(data_body[0]))
```

```python
import pandas as pd
```

```python
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
soup = BeautifulSoup(str(data_body[0]), 'html.parser')
data = []
for row in soup.find_all('tr'):
    values = [td.text if td.text.strip() != '' else np.nan for td in row.find_all('td')]
    data.append(values)
columns = ['Year', 'Jan' , 'Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec','Annual']
df = pd.DataFrame(data, columns=columns)
for col in columns[1:]:
    df[col] = pd.to_numeric(df[col], errors='coerce')

print(df)
```

```python
main_path = '/content/drive/MyDrive/projects_data/FYP_20_Erandi'
```

```python
df.to_csv(f"{main_path}/german_cpi_data.csv")
```
